```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
plot_y = (plot_x-2.5)**2-1

plt.plot(plot_x, plot_y)
plt.show()
```

## 对一个函数（损失函数），在某个任意点上，通过求导可以得到其斜率，如果斜率为正，说明该函数向前是会增加，应当向后，移动方向为减去斜率*学习率，如果斜率是负的，则相反


## 损失函数J

```python
def J(theta):
    return (theta-2.5)**2-1
```

## 损失函数J的导数

```python
def dJ(theta):
    return 2*(theta-2.5)
```

## 超参数学习率、迭代次数

```python
eta = 0.1
epsilon = 1e-8
theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    if (abs(J(theta) - J(last_theta)) < epsilon):
        break
print(theta)
print(J(theta))
```

## 针对局部最优解，多次运行并随机化初始点
初始点位置也是超参数

```python
theta = 0.0
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    if (abs(J(theta) - J(last_theta)) < epsilon):
        break
plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
plt.show()
```
